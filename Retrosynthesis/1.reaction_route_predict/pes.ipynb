```python
from rdkit import Chem
import pandas as pd
```

```python
def check_error(smiles_file):
    with open(smiles_file,'r') as f:
        smiles = f.readline().rstrip()
        err_smiles = list()
        save_smiles = list()
        while smiles:
            try:
                mol = Chem.MolFromSmiles(smiles)
                Chem.SanitizeMol(mol)
                save_smiles.append(smiles + '\n')
            except:
                err_smiles.append(smiles)
            smiles = f.readline().rstrip()

    # save error smiles
    if err_smiles:
        err_df = pd.DataFrame(err_smiles,columns=['smiles'])
        err_df.to_csv('err_smiles.csv',index=None)
        print(err_smiles)
    # save check smiles
    if save_smiles:
        checked_file = 'checked_smi.smi'
        with open(checked_file,'w') as f:
            f.writelines(save_smiles)
smiles_file = 'Pesticide.csv'
check_error(smiles_file)
```

```python
import os,subprocess,shutil,re,json
```

```python
def predict_route(dir):
    files = os.listdir(dir)
    patt = r'drug_(\d+)\.csv'
    use_csv = list()
    for file in files:
        if re.search(patt,file):
            use_csv.append(file)
    use_csv.sort(key=lambda s:int(re.search(patt,s).group(1)))
    
    print(use_csv)
    i =0
    for file in use_csv:
        correct_file = '/home/zhouhao/zzl/price_predict/test2/usr.json'
        content = dict()
        content["smiles_file"] = '/home/work/store/' + file
        with open(correct_file,'w') as f:
            json.dump(content,f)
        
        args = ['docker','run','--rm','--gpus','all','-v','./test2:/home/work/store','reaction_route:20240424','/home/work/store/usr.json','/home/work/store']
        subprocess.run(args)
        result_file = '/home/zhouhao/zzl/price_predict/test2/output.json'
        correct_result = f'/home/zhouhao/zzl/price_predict/test2/output{i}.json'
        os.rename(result_file,correct_result)
        i += 1

        files1 = os.listdir(dir)
        patt1 = '.png'
        remove_files = list()
        for file in files1:
            if patt1 in file:
                n = os.path.join(dir,file)
                remove_files.append(n)
        for file in remove_files:
            os.remove(file)
```
